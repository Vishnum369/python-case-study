```python
import numpy as np
import pandas as pd

df = pd.read_csv("/content/sales-data.csv.txt")
```

```python
df
```

```python
df.head()
```

```python
df['OrderDate'] = pd.to_datetime(df['OrderDate'])
df.info()
```

```python
df['Totalprice'] = df['Quantity'] * df['UnitPrice'] * (1-df['Discount'])
df.head()
```

```python
def badge(data):
  if data >= 15:
    return 'Platinum'
  elif data >= 10:
    return 'Gold'
  elif data > 5:
    return 'Silver'
  else:
    return 'Bronze'
```

```python
df['badge'] = df.Totalprice.apply(badge)
df
```

```python
import sqlite3
conn = sqlite3.connect('sales.db')

df.to_sql('sales', conn, index=False, if_exists='replace')
```

```python
cc_df = pd.read_sql_query("SELECT * FROM sales WHERE PaymentMethod = 'Credit Card'", conn)
cc_df
```

```python
g_df = pd.read_sql_query("SELECT * FROM sales WHERE badge = 'Gold'", conn)
g_df
```

```python
p_df = pd.read_sql_query("SELECT * FROM sales WHERE badge = 'Silver'",conn)
p_df
```

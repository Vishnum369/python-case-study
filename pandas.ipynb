```python
import seaborn as sns
data = sns.load_dataset("titanic")
data.head(10)
```

```python
data.isnull().sum()
```

```python
data.groupby('sex')['survived'].mean()*100
```

```python
print("Average age:",data['age'].mean())
print("Median age:",data['age'].median())
print("Standard Deviation of age:",data['age'].std())
```

```python
print(data['pclass'].value_counts())
```

```python
female_young_survived=data[(data['sex']=='female') & (data['age']<18) & (data['survived']==1)]
print(female_young_survived)
```

```python

median_age = data['age'].median()
print("Median age:", median_age)


data['age'] = data['age'].fillna(median_age)

print("Missing age count:", data['age'].isnull().sum())
```

```python
import numpy as np

data['age_group'] = np.where(data['age'] < 18, 'Child',
                    np.where(data['age'] <= 60, 'Adult', 'Senior'))
print(data[['age', 'age_group']].head(10))
```

```python
data.nlargest(5,'fare')
```

```python
data.groupby(['pclass','sex'])['fare'].mean()
```

```python
data.rename(columns={'sex':'gender','fare':'ticket_fare'},inplace=True)
```

```python
print(data.groupby(['pclass','survived']).size())
```

```python
print(data.groupby(['pclass','gender'])['age'].mean())
```

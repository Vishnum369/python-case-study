```python
import pandas as pd
import sqlite3
```

```python
#load data
url = "https://earthquake.usgs.gov/earthquakes/feed/v1.0/summary/all_month.csv"
df = pd.read_csv(url)
df
```

```python
print("Number of records:", len(df))
print("Unique locations:", df['place'].nunique())
```

```python
print("Top 5 rows:")
print(df.head())
print("Column names:")
print(df.columns)
```

```python
# convert time to datetime
df['time'] = pd.to_datetime(df['time'])
df['time']
```

```python
#drop records with missing latitude,longitude,mag
df = df.dropna(subset=['latitude', 'longitude', 'mag'])
df
```

```python
#filter mag
df = df[df['mag'] >= 4.0]
df
```

```python
#add day of week
df['day_of_week'] = df['time'].dt.day_name()
df['day_of_week']
```

```python
#add security level
def severity(mag):
    if mag < 4.0:
        return "Low"
    elif mag <= 6.0:
        return "Moderate"
    else:
        return "High"

df['severity_level'] = df['mag'].apply(severity)
df['severity_level']
```

```python
#count earthquick per place
counts = df.groupby('place').size().reset_index(name='earthquake_count')
counts
```

```python
#avg magnitude and max depth per day
df['date'] = df['time'].dt.date
summary = df.groupby('date').agg({
    'mag': 'mean',
    'depth': 'max'
}).reset_index().rename(columns={'mag': 'avg_magnitude', 'depth': 'max_depth'})
summary
```

```python
#save cleaned data
df.to_csv('cleaned_earthquakes.csv', index=False)
counts.to_csv('earthquake_counts.csv', index=False)
summary.to_csv('earthquake_summary.csv', index=False)
```

```python
#store to sqlite
conn = sqlite3.connect('earthquakes.db')
df.to_sql('earthquakes', conn, if_exists='replace', index=False)
counts.to_sql('earthquake_counts', conn, if_exists='replace', index=False)
summary.to_sql('earthquake_summary', conn, if_exists='replace', index=False)
conn.close()

print("ETL task completed successfully!")
```
